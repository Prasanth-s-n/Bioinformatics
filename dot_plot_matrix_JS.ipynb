```javascript
function subConsarr(arr){
    var retarr=[]
    var S = []
    var s=0
    for(var i=0;i<arr.length-1;i++){
        if (arr[i]==arr[i+1]-1){
            s=s+1;
        }
        else{
            retarr.push(arr.slice(i-s,i+1))
            s=0
        }
        if (i==arr.length-2){
            if (arr[arr.length-1]==arr[arr.length-2]+1){
                retarr.push(arr.slice(i-s+1,i+2))
            }
            else{
                retarr.push(arr.slice(i-s,i+1))
            }
        }   
    }
    for(var ra = 0;ra<retarr.length;ra++){
        if (retarr[ra].length!=1){
            S.push(retarr[ra])
        }
    }
    return S
}
```

```javascript
function getDiag(matrix,iod){
    var tiR=0
    var tiC=0
    if (iod<0){
        tiR=math.abs(iod)
    }
    else{
        tiC=iod
    }
    var diag = []
    var nRow = matrix.length
    var nCol = matrix[0].length
    var done = true
    while (done){
        if ((tiR<nRow) & (tiC<nCol)){ 
            diag.push(matrix[tiR][tiC])
            tiR = tiR+1
            tiC = tiC+1
        }
        else{done=false}
    }
    return diag
}

function original_index(matrix,iod){
    
    var tiR=0
    var tiC=0
    if (iod<0){
        tiR=math.abs(iod)
    }
    else{
        tiC=iod
    }
    
    var diag = []
    var nRow = matrix.length
    var nCol = matrix[0].length
    var done = true
    
    while (done){
        if ((tiR<nRow) & (tiC<nCol)){ 
            
            diag.push([tiR,tiC])
            tiR = tiR+1
            tiC = tiC+1
        }
        else{done=false}
    }
    return diag
}
```

```javascript
function getAllIndexes(arr, val) {
    var indexes = [], i = -1;
    while ((i = arr.indexOf(val, i+1)) != -1){
        indexes.push(i);
    }
    return indexes;
}
```

```javascript
function subSequences(dpm,s1,s2){
    
    var cdpm = dpm
    var retSeq = []
    t = []
    var nRow = dpm.length
    var nCol = dpm[0].length
    for(var i=-nRow+1;i<nCol;i++){
        
        curr_diag = getDiag(dpm,i)
        subarrs = subConsarr(getAllIndexes(curr_diag,1))
        for(var j=0;j<subarrs.length;j++){
            var sa = subarrs[j]
            var diag_original_index = original_index(dpm,i)
            var seqRet = ""
            for(var k=0;k<sa.length;k++){
                seqRet = seqRet+s2[diag_original_index[sa[k]][0]]
            }
            retSeq.push(seqRet)
        }
    }
    
    return retSeq
}
```

```javascript
function tscore(s1,s2){
    
    var l = s1.length
    var match = 0
    var mismatch = 0
    var gap = 0
    for(var i=0;i<l;i++){
        if (s1[i]==s2[i]){
            match+=1}
        else if (s1[i]=="-" || s2[i]=="-"){
            gap+=1}
        else{
            mismatch+=1}
    }
    var tscore=2*match + -1*(mismatch) + -2*gap
    return tscore
    
}
```
